## **Remove the row with the single outlier in num_accts. Filter with a logical condition**

```python
# Remove the row with the single outlier in num_accts. Filter with a logical condition.


df.sort_values(['num_accts'],ascending =True).tail()    # With this we found out that 3345 value which can be considered as outlier.
        


```

```python
df = df.drop(df[df['num_accts'] > 4].index)  
 
df.tail()           # Outlier is removed.
```

## **Remove the rows with negative income values. Filter with a logical condition**

```python
# Remove the rows with negative income values. Filter with a logical condition.


df = df.drop(df[df['income'] < 0].index) 

df.tail()
```
